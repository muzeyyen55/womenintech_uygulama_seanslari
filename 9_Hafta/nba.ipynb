```python
import pandas as pd
import numpy as np

# Verileri okuyalÄ±m.
games_details = pd.read_csv("nba_data/games_details.csv")
games = pd.read_csv("nba_data/games.csv")
players = pd.read_csv("nba_data/players.csv")
ranking = pd.read_csv("nba_data/ranking.csv")
teams = pd.read_csv("nba_data/teams.csv")

# Verilerin iÃ§erisinde bulunan satÄ±rlarÄ± inceleyelim.


games.rename(columns= {'GAME_DATE_EST':'Oyun_Tarihi','GAME_ID': 'Oyun_ID', 'GAME_STATUS_TEXT':'Mac_Turu',
              'HOME_TEAM_ID':'Ev_Takimi_Id_1','VISITOR_TEAM_ID': 'Deplasman_Takim_Id', 'SEASON':'Sezon',
              'TEAM_ID_home':'Ev_Takimi_Id','PTS_home': 'Ev_Takimi_Sayisi', 'FG_PCT_home':'Ev_Sut_Isabet_Yuzdesi',
              'FT_PCT_home':'Ev_Serbest_Atis_Isabet_Yuzdesi','FG3_PCT_home': 'Ev_3luk_Isabet_Yuzdesi', 'AST_home':'Ev_Asist_Sayisi',
              'REB_home':'Ev_Ribaund_Sayisi','TEAM_ID_away':'Deplasman_Takimi_Id','PTS_away': 'Deplasman_Takimi_Sayisi', 'FG_PCT_away':'Deplasman_Sut_Isabet_Yuzdesi',
              'FT_PCT_away':'Deplasman_Serbest_Atis_Isabet_Yuzdesi','FG3_PCT_away': 'Deplasman_3luk_Isabet_Yuzdesi', 'AST_away':'Deplasman_Asist_Sayisi',
              'REB_away':'Deplasman_Ribaund_Sayisi','HOME_TEAM_WINS':'Ev_Sahibi_Kazandi'
            },inplace = True)

games.dropna(subset=["Oyun_ID"],inplace = True)# OyunId si null olan degerler silindi. OyunIdsi hic null olmadigi icin herhangi bir veri silinmedi.
# games.dropna(inplace = True)# Satirda herhangi bir deger null ise o satir silindi.
# games.fillna(0,inplace= True)# Bos satirlari silmek yerine onlari doldurabiliriz de.
# games.fillna(games.mean(), inplace= True)# 0 ile doldurmak yerine ortalama ile de doldurabiliriz.
games['Ev_Takimi_Sayisi'].fillna(games['Ev_Takimi_Sayisi'].mean(),inplace = True)
# print(games[(games['Oyun_ID'] == 10300116)])

games.drop(['Ev_Takimi_Id_1'], axis= 1, inplace= True) #Ev takiminin idsi 2 kolonda bulundugu icin bunlardan birini kaldiralim.

cleanGames = games[['Ev_Takimi_Id','Ev_Takimi_Sayisi','Sezon']]



print(cleanGames.groupby(['Ev_Takimi_Id', 'Sezon']).value_counts())
print(cleanGames.groupby(['Ev_Takimi_Id', 'Sezon']).agg({'Ev_Takimi_Sayisi': ['mean']}))



# checkDataFrame(games_details, 'game details')
# checkDataFrame(players, 'players')
# checkDataFrame(ranking, 'ranking')
# checkDataFrame(teams, 'teams')
```
